# Oefening: Cross-validatie

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Lees de gegevens
train_data = pd.read_csv('../data/train.csv', index_col='Id')
test_data = pd.read_csv('../data/test.csv', index_col='Id')

# Verwijder rijen met ontbrekend doel, scheid doel van voorspellers
train_data.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = train_data.SalePrice
train_data.drop(['SalePrice'], axis=1, inplace=True)

# Selecteer alleen numerieke kolommen
numerieke_kolommen = [cname for cname in train_data.columns if train_data[cname].dtype in ['int64', 'float64']]
X = train_data[numerieke_kolommen].copy()
X_test = test_data[numerieke_kolommen].copy()
```

```python
X.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

mijn_pijpleiding = Pipeline(steps=[
   ('preprocessor', SimpleImputer()),
   ('model', RandomForestRegressor(n_estimators=50, random_state=0))
   ])
```

```python
from sklearn.model_selection import cross_val_score

# Vermenigvuldig met -1 omdat sklearn *negatieve* MAE berekent
scores = -1 * cross_val_score(mijn_pijpleiding, X, y, cv=5, scoring='neg_mean_absolute_error')

print("Gemiddelde MAE-score:", scores.mean())
```
